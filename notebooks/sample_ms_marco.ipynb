```python
import ir_datasets
import pandas as pd


dataset = ir_datasets.load("msmarco-document/train")
dataset.docs_count()
```

```python
queries_sample = []

for query in dataset.queries_iter()[::10]: # Хотим в 10 раз меньше
    queries_sample.append((query.query_id, query.text))
```

```python
query_ids_sampled = set([x[0] for x in queries_sample])
doc_ids_for_queries = []
qrels_sampled = []


for qrel in dataset.qrels_iter(): # Семплим все документы релевантные и не очень к запросу
    if qrel.query_id in query_ids_sampled:
        doc_ids_for_queries.append(qrel.doc_id)
        qrels_sampled.append((qrel))
```

```python
len(queries_sample), len(qrels_sampled)
```

```python
qrels = pd.DataFrame(qrels_sampled)
```

```python
doc_ids = set(qrels.doc_id)
```

```python
from tqdm import tqdm
documents_train_sample = []

for doc in tqdm(dataset.docs_iter()):
    if doc.doc_id in doc_ids:
        documents_train_sample.append(doc)

train_documents = pd.DataFrame(documents_train_sample)
train_documents.to_csv("datasets/documents_train.csv")
```
